```python
import csv



fileName = "ExpensesTracker"
user_selected_option = - 1
user_expeses = []


#function to write the expenses in file 
def writeExpenseInFile():
    file = open(fileName, "w")
    file.write(str(user_expeses))
    file.close()


#function to read the expenses in file 
def readExpenseFromFile():
    try:
      file = open(fileName, "r")
      return eval(file.read())
    except FileNotFoundError:
      print ("There is no expenses recorded so far")  
      return []

    
#function to read the expenses in file 
# If any required details (date, category, amount, or description) are missing, notifying the user that itâ€™s incomplete
def checkIfExpensesValid(expense):
    message = ""
    if len(expense["date"]) == 0:
        message = message + "\nDate"
    if len(expense["category"]) == 0:
        message = message + "\nCategory"
    if len(str(expense["amount"])) == 0:
        message = message + "\nAmount"
    if len(expense["description"]) == 0:
        message = message + "\nDescription"

    if len(message) == 0:
        # it means there no mesaseg so far, so no misisng entry - we can return true 
        return (True, "")
    else: 
         return (False, "Below Entries are misisng for expense: " + message)

        
#function to calculate the expenses so far 
def calculateExpnseSoFar():
     total_expenses = 0
     expenses = readExpenseFromFile()      
     for expense in expenses:
        total_expenses += total_expenses + int(expense["amount"])
     return total_expenses


#function to save the data in csv file 
def saveExpenses():
    expenses = readExpenseFromFile()
    if len(expenses) > 0: 
      file = open("data.csv", "w")
      writer = csv.DictWriter(file, fieldnames = expenses[0].keys())
      writer.writeheader()
      writer.writerows(expenses)
      print("Records are exported to data.csv file.")
    else:
        print("There is no expense to save in CSV")
    

def addExpenses():
    date = input("Enter the date of the expense in the format YYYY-MM-DD: ")
    category = input("Enter the category of the expense, such as Food or Travel: ")
    amount = float(input("Enter the amount spent: "))
    description = input("Enter the brief description of the expense: ")
    user_expeses.append({"date":date, "category": category, "amount": amount, "description":description}) 
    writeExpenseInFile()

def viewExpenses():
    expenses = readExpenseFromFile()
    for expense in expenses:
        isValidExpenses = checkIfExpensesValid(expense)
        if isValidExpenses[0]:
            print (str(expense))
        else:
           print(isValidExpenses[1])
            
def trackBudget():
     budget = float(input("Enter the total amount you want to budget for the month: "))
     total_expenses_so_far = calculateExpnseSoFar()
     if budget - total_expenses_so_far > 0:
         print("You have ", str(budget - total_expenses_so_far) , "left for the month")
     else:
         print("You have exceeded your budget!")

    
def getThankYouMessage(selected_option):
    return "Thank you for " + selected_option + ". please choose another option to move forward.\n"


while user_selected_option != 5:
    print("\n")
    print("Welcome to expenses tracker manager")
    print("1.Add expense")
    print("2.View expenses")
    print("3.Track budget")
    print("4.Save expenses")
    print("5.Exit")
    print("\n")

    user_selected_option = int(input("Please enter your choice: "))

    match user_selected_option:
        case 1:
            addExpenses()
            print(getThankYouMessage("Adding expenses")) 
        case 2:
            viewExpenses()
        case 3:
            trackBudget()
        case 4:
            saveExpenses()
        case 5:
            saveExpenses()
            print("Thank you for using the tracker manager") 
            

                               
                           
```
